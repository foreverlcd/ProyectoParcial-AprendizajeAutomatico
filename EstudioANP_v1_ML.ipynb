# ESTUDIO DE LOS FACTORES DETERMINANTES EN LA FRECUENCIA DE VISITAS A LAS ÁREAS NATURALES PROTEGIDAS DEL PERÚ

## Problema

A pesar de la relevancia del turismo en las ANP, existen brechas en el conocimiento sobre los factores que más influyen en la frecuencia.  La falta de estudios limita la capacidad de los gestores de ANP para tomar decisiones informadas y mejorar las estrategias de planificación y promoción turística.

## Preguntas de Investigación

¿Cuáles son los principales factores que influyen en la frecuencia de visitas a las Áreas Naturales Protegidas del Perú?

## Objetivo General

Identificar los factores clave que influyen en la frecuencia de las visitas turísticas en las Áreas Naturales Protegidas del Perú, utilizando los datos proporcionados por SERNANP.

## Justificación e Importancia del Estudio
El presente estudio es relevante debido a la creciente presión sobre las Áreas Naturales Protegidas por el aumento del turismo. A través de un análisis robusto de los datos de visitas, el estudio ayudará a identificar patrones que pueden guiar la formulación de políticas de gestión sostenible, lo que será útil para los gestores de ANP, operadores turísticos y responsables de políticas públicas. Este trabajo contribuirá al desarrollo de estrategias que equilibren el turismo y la conservación de los recursos naturales.

### Transformación de variables

```python
# -- Transformar la variable SECTOR a NUMERO_SECTORES

# Contar sectores por ANP, excluyendo la etiqueta 'NINGUNO'
sectores_por_anp = data[data['SECTOR'] != 'NINGUNO'].groupby('ANP')['SECTOR'].nunique().reset_index()
sectores_por_anp.columns = ['ANP', 'NUMERO_SECTORES']

# Combina el conteo de sectores con el DataFrame original
data = data.merge(sectores_por_anp, on='ANP', how='left')

# Rellenar con 0 los ANP que tienen 'NINGUNO' como sector
data['NUMERO_SECTORES'].fillna(0, inplace=True)

# Transformar la columna 'columna_float' de float a int
data['NUMERO_SECTORES'] = data['NUMERO_SECTORES'].astype(int)

# Elimina la columna 'SECTOR'
data.drop(columns=['SECTOR'], inplace=True)
```

```python
def imputar_outliers(df, columna, reemplazo):
    # Calcular los cuartiles
    Q1 = df[columna].quantile(0.25)
    Q3 = df[columna].quantile(0.75)
    IQR = Q3 - Q1

    # Definir los límites inferior y superior
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR

    # Identificar los valores atípicos
    outliers = df[(df[columna] < limite_inferior) | (df[columna] > limite_superior)]

    # Mostrar los valores atípicos identificados
    if not outliers.empty:
        outliers_unicos = outliers[columna].unique()
        print(f"Valores atípicos identificados: {outliers_unicos}")
        print(f"Cantidad de valores atípicos: {len(outliers)}")
        print(f"Porcentaje de valores atípicos sobre el total: {len(outliers) / len(df) * 100:.2f}%")

    # Realizar la imputación o eliminación según el parámetro 'reemplazo'
    if reemplazo == 'L':
        # Reemplazar los outliers con los límites
        df[columna] = df[columna].apply(
            lambda x: limite_inferior if x < limite_inferior else (
                limite_superior if x > limite_superior else x
            )
        )
    elif reemplazo == 'M':
        # Calcular la media de la columna
        media = df[columna].mean()
        # Reemplazar los outliers con la media
        df[columna] = df[columna].apply(
            lambda x: media if x < limite_inferior or x > limite_superior else x
        )
    elif reemplazo == 'N':
        # Calcular la mediana de la columna
        mediana = df[columna].median()
        # Reemplazar los outliers con la mediana
        df[columna] = df[columna].apply(
            lambda x: mediana if x < limite_inferior or x > limite_superior else x
        )
    else:
        # Eliminar los outliers
        df = df[(df[columna] >= limite_inferior) & (df[columna] <= limite_superior)]

    return df
```

## Selección de Modelo

```python
# Define las características (X) y el target (y)
X = data.drop(columns=['VISITAS'])
y = data['VISITAS']

# Dividir los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=100, shuffle=True, test_size=0.2)

# Mostrar las formas de los conjuntos de entrenamiento y prueba
print(f"Forma de X_train: {X_train.shape}")
print(f"Forma de X_test: {X_test.shape}")
print(f"Forma de y_train: {y_train.shape}")
print(f"Forma de y_test: {y_test.shape}")
```

```python
# Clase para preprocesamiento y escalado
class DataPreprocessing:
    ssc = StandardScaler()  # Instancia de StandardScaler

    @staticmethod
    def scaling(X: pd.DataFrame, data_type: str) -> pd.DataFrame:
        """Escalado de los datos (standardization) para modelos lineales."""
        if data_type.lower() == 'train':
            return pd.DataFrame(DataPreprocessing.ssc.fit_transform(X), columns=X.columns, index=X.index)
        if data_type.lower() == 'test':
            return pd.DataFrame(DataPreprocessing.ssc.transform(X), columns=X.columns, index=X.index)

# Escalado de datos para los modelos lineales (entrenamiento y prueba)
X_train_scale = DataPreprocessing.scaling(X_train, 'train')
X_test_scale = DataPreprocessing.scaling(X_test, 'test')
```

```python
X_train_scale.describe()
```

```python
X_test_scale.describe()
```

```python
# DataFrame para almacenar los resultados de los modelos
df_models = pd.DataFrame(columns=['Algorithm', 'r2_train', 'r2_test'])

def make_model(X_tr, X_te, y_tr, y_te, model, model_name: str):
    """Entrenar el modelo y calcular los r2 scores"""
    model.fit(X_tr, y_tr)
    y_pred_train = model.predict(X_tr)
    y_pred_test = model.predict(X_te)
    r2_train = r2_score(y_tr, y_pred_train)
    r2_test = r2_score(y_te, y_pred_test)
    df_models.loc[len(df_models)] = [model_name, r2_train, r2_test]

# Modelos para probar
models = [
    (LinearRegression(), 'LinearRegression'),
    (GradientBoostingRegressor(), 'GradientBoosting'),
    (RandomForestRegressor(), 'RandomForest')
]

# Aplicar cada modelo y calcular los resultados
for model, model_name in models:
    if model_name in ['LinearRegression', 'Ridge', 'Lasso', 'ElasticNet']:
        # Usar los datos escalados para modelos que necesitan estandarización
        make_model(X_train_scale, X_test_scale, y_train, y_test, model, model_name)
    else:
        # Usar los datos originales para modelos que no necesitan estandarización
        make_model(X_train, X_test, y_train, y_test, model, model_name)

# Mostrar los resultados de los modelos
print(df_models)
```

```python
# Crear el gráfico de barras comparando R2 train y test
fig, ax = plt.subplots(figsize=(10, 6))

# Ancho de las barras
bar_width = 0.35

# Posiciones de las barras en el eje x
indices = np.arange(len(df_models['Algorithm']))

# Barras para R2 train
rects1 = ax.bar(indices, df_models['r2_train'], bar_width, label='R2 Train', color='#68cf00')

# Barras para R2 test
rects2 = ax.bar(indices + bar_width, df_models['r2_test'], bar_width, label='R2 Test', color='#054c2c')

# Añadir etiquetas, título y leyendas
ax.set_xlabel('Algoritmo')
ax.set_ylabel('Coeficiente R2')
ax.set_title('Comparación de R2 en conjuntos de entrenamiento y prueba por algoritmo')
ax.set_xticks(indices + bar_width / 2)
ax.set_xticklabels(df_models['Algorithm'])
ax.legend()

# Función para añadir etiquetas de valor en las barras
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('%.3f' % height,
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

# Añadir las etiquetas de valor a las barras
autolabel(rects1)
autolabel(rects2)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

## Ajuste de Hiperparámetros

```python
# Dividir en entrenamiento y validación
X_trainval, X_valid, y_trainval, y_valid = train_test_split(X_train, y_train, shuffle=True,  random_state=100)
```

```python
# Definir la función objetivo para RandomForestRegressor
def objective_rf(trial):
    # Espacio de búsqueda para los hiperparámetros de RandomForestRegressor
    param = {
        'n_estimators': trial.suggest_int('n_estimators', 100, 1000),
        'max_depth': trial.suggest_int('max_depth', 2, 32, log=True),
        'min_samples_split': trial.suggest_int('min_samples_split', 2, 20),
        'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 20),
        'bootstrap': trial.suggest_categorical('bootstrap', [True, False]),
        'random_state': 100
    }

    # Entrenar modelo RandomForestRegressor
    model = RandomForestRegressor(**param)
    model.fit(X_trainval, y_trainval)

    # Predicción y cálculo de métrica (R2)
    preds = model.predict(X_valid)
    r2 = r2_score(y_valid, preds)

    return r2
```

```python
# Crear un estudio Optuna y optimizar
study_rf = optuna.create_study(direction='maximize')
study_rf.optimize(objective_rf, n_trials=100)

print("Mejores hiperparámetros para RandomForestRegressor:", study_rf.best_params)
print("Mejor R2 score para RandomForestRegressor:", study_rf.best_value)
```

```python
# Historial de optimización para el estudio de RandomForestRegressor
plot_optimization_history(study_rf).show()

# Importancia de los hiperparámetros para el estudio de RandomForestRegressor
plot_param_importances(study_rf).show()
```

```python
# Entrenar modelo final de RandomForestRegressor con los mejores hiperparámetros
# Obtener los mejores hiperparámetros del estudio de RandomForestRegressor
best_params_rf = study_rf.best_params

# Crear una instancia de RandomForestRegressor con los mejores hiperparámetros
best_model_rf = RandomForestRegressor(**best_params_rf, random_state=100) # Agrega random_state para reproducibilidad

# Entrenar el modelo final con el conjunto de entrenamiento completo
best_model_rf.fit(X_train, y_train)
```

```python
# Definir la función objetivo para GradientBoosting
def objective_gradient_boosting(trial):
    # Espacio de búsqueda para los hiperparámetros
    param = {
        'n_estimators': trial.suggest_int('n_estimators', 100, 1000),
        'learning_rate': trial.suggest_loguniform('learning_rate', 0.001, 0.1),
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'subsample': trial.suggest_float('subsample', 0.5, 1.0, step=0.1),
        'min_samples_split': trial.suggest_int('min_samples_split', 2, 20),
        'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 20),
        'random_state': 100
    }

    # Entrenar modelo GradientBoostingRegressor
    model = GradientBoostingRegressor(**param)
    model.fit(X_trainval, y_trainval)

    # Predicción y cálculo de métrica
    preds = model.predict(X_valid)
    r2 = r2_score(y_valid, preds)

    return r2
```
